## Numerical Differentiation
### First Order Differentiation

```python
def forward_difference(f, x, h):
    return (f(x+h) - f(x))/h

def backward_difference(f, x, h):
    return (f(x) - f(x-h))/h

def central_difference(f, x, h):
    return (f(x+h) - f(x-h))/(2*h)
```

```python
def func(x):
    return 1/(1 + np.exp(-x))   ## Sigmoid function
```

```python
def exact(x):
    return func(x) - func(x)**2  ## differentiation of sigmoid function
```

```python
x = 10; h = 0.1
print("Forward Difference = ", forward_difference(func, x, h))
print("Backward Difference = ", backward_difference(func, x, h))
print("Central Difference = ", central_difference(func, x, h))
print("Exact value = ",exact(x))
```

```python
h = [0.1,0.05,0.01,0.005,0.001,0.0005,0.0001]
forward_error = []
backward_error = []
central_error = []
```

```python
for dx in h:
    forward_error.append(abs(forward_difference(func, x, dx) - exact(x)))
    backward_error.append(abs(backward_difference(func, x, dx) - exact(x)))
    central_error.append(abs(central_difference(func, x, dx) - exact(x)))
```

```python
plt.figure(figsize=(8,6))
plt.plot(np.log(h), forward_error, 'red', label='forward')
plt.plot(np.log(h), backward_error, 'blue', label='backward')
plt.plot(np.log(h), central_error, 'black', label='central')
plt.legend()
plt.xlabel('log(h)')
plt.ylabel('error')
plt.title("Error vs log(h)")
plt.show()
```

### Second Order Differentiation

```python
def forwardDifference(f, x, h):
    return (f(x+h+h) - 2*f(x+h) + f(x))/h**2

def backwardDifference(f, x, h):
    return (f(x) - 2*f(x-h) + f(x-h-h))/h**2

def centralDifference(f, x, h):
    return (f(x+h) - 2*f(x) + f(x-h))/h**2

def myFunc(x):
    return 2*x**3 + 4*x**2 - 5*x              ### example function

x = 1
h = 0.1

dydt_f = forwardDifference(myFunc, x, h)
dydt_b = backwardDifference(myFunc, x, h)
dydt_c = centralDifference(myFunc, x, h)

def exact(x): 
    return 12*x + 8                             ### second order differential equation of the function
 
print("Forward difference =", dydt_f)
print("Backward difference =", dydt_b)
print("Central difference =", dydt_c)
print("Exact value =", exact(x))
```

```python
h = [0.1,0.05,0.01,0.005,0.001,0.0005,0.0001]
forward_error = []
backward_error = []
central_error = []
```

```python
for dx in h:
    forward_error.append(abs(forwardDifference(func, x, dx) - exact(x)))
    backward_error.append(abs(backwardDifference(func, x, dx) - exact(x)))
    central_error.append(abs(centralDifference(func, x, dx) - exact(x)))
```

```python
plt.figure(figsize=(8,6))
plt.plot(np.log(h), forward_error, 'red', label='forward')
plt.plot(np.log(h), backward_error, 'blue', label='backward')
plt.plot(np.log(h), central_error, 'black', label='central')
plt.legend()
plt.xlabel('log(h)')
plt.ylabel('error')
plt.title("Error vs log(h)")
plt.show()
```

## Integration by Trapezoidal Rule

```python
def trapz(f, a, b, n=50):
    x = np.linspace(a,b,n+1)   ## we need n equal sub-divisons
    y = f(x)
    y_l = y[:-1]               ## exclude the rigth end
    y_r = y[1:]                ## exclude the left end
    h = (b - a)/n
    integral = (h/2)*np.sum(y_l + y_r)
    return integral
```

## Polynomial Expansion of functions

```python
f = lambda x: np.sin(x)
x = np.pi/2
a0 = f(0)                         ## y = sin(x)    first term
a1 = np.cos(0)*x                  ## y1 = cos(x)   second term        
a2 = np.power(x,2)*(-np.sin(0))/2 ## y2 = -sin(x)  third term
a3 = np.power(x,3)*(-np.cos(0))/6 ## y3 = -cos(x)  fourth term
```

```python
f = lambda x: np.exp(-x)
x = 0
a0 = f(0)                         ## y = e^(-x)
a1 = x*(-f(0))                    ## y1 = -e^(-x)          
a2 = np.power(x,2)*(f(0))/2       ## y2 = e^(-x)
a3 = np.power(x,3)*(-f(0))/6      ## y3 = -e^(-x)
```

```python
error_ls = []
error_ls.append(abs(f(x) - a0))
error_ls.append(abs(f(x) - (a0+a1)))
error_ls.append(abs(f(x) - (a0+a1+a2)))
error_ls.append(abs(f(x) - (a0+a1+a2+a3)))
```

```python
plt.figure(figsize=(8,6))
plt.plot(error_ls)
plt.grid()
plt.xlabel('No. of terms in polynomial expansion')
plt.ylabel('error')
plt.title('Talor Series error graph')
plt.show()
```

## Partial Derivatives

```python
f = lambda x1,x2: x1 - x2 + 2*x1*x2 + 2*x1**2 + x2**2

x1 = np.linspace(-10,10,100)
x2 = np.linspace(-10,10,100)
```

```python
def partial_derivative_2d(f, x, y, h):
    dfdx = (f(x+h,y) - f(x-h,y))/(2*h)    ## central difference
    dfdy = (f(x,y+h) - f(x,y-h))/(2*h)
    return np.array([dfdx,dfdy])
```

```python
h = 0.01
x, y = [1], [5]
grad = []
for x,y in zip(x,y):
    grad.append(partial_derivative_2d(f,x,y,h))
grad
```

```python
def contour_plot_with_grad(f,x,y,grad):
    X, Y = np.meshgrid(x, y)
    z = f(X,Y)

    fig, ax = plt.subplots(figsize=(8,6))
    cp = ax.contour(X, Y, z)
    fig.colorbar(cp) # Add a colorbar to a plot
    ax.set_title('Contour Plot with grad arrow')
    for g in grad:
        ax.arrow(0,0,g[0],g[1],head_length=0.2,head_width=0.1,color='red',length_includes_head=True)
    plt.show()
```

## Application

```python
## function to convert a coloured image to gray scale
def colour_to_gray(img_path):
    img = plt.imread(img_path)
    height, width, channels = img.shape

    img_gray = np.zeros((height,width))   ## placeholder for the gray image

    for i in range(height):
        for j in range(width):
            img_gray[i,j] = np.sum(img[i,j])/channels  ## mean value of the pixel

    img_gray = img_gray.astype(np.uint8)
    return img_gray
```

```python
img_path = "../week2/coloured_image.jpeg"
img = plt.imread(img_path)
img_gray = colour_to_gray(img_path)

fig, ax = plt.subplots(1,2,figsize=(12,10))
ax[0].imshow(img)
ax[0].set_title("Original Image")
ax[1].imshow(img_gray,cmap="gray")
ax[1].set_title("Gray-scaled Image")
plt.show()
```

## Edge Detection

```python
def edge_detection(img, grad_threshold):
    height,width = img.shape
    img_with_edge = np.zeros(img.shape)

    for i in range(height-1):
        for j in range(width-1):
            grad = abs(int(img[i,j+1]) - int(img[i,j]))  ## grad calculation along the width
            if grad >= grad_threshold:
                img_with_edge[i,j] = 255
            else:
                img_with_edge[i,j] = 0

    return(img_with_edge.astype(np.uint8))
```

```python
img_path = "../week2/subject01.sad.pgm"
grad_threshold = 20
img = plt.imread(img_path)
img_edge = edge_detection(img, grad_threshold)

fig, ax = plt.subplots(1,2,figsize=(12,10))
ax[0].imshow(img,cmap="gray")
ax[0].set_title("Original Image")
ax[1].imshow(img_edge,cmap="gray")
ax[1].set_title("Edge-detected Image")
plt.show()
```

```python
img_path = "../week2/coloured_image.jpeg"
img = plt.imread(img_path)
img_gray = colour_to_gray(img_path)
grad_threshold = 30
img_edge = edge_detection(img_gray, grad_threshold)

fig, ax = plt.subplots(1,3,figsize=(16,12))
ax[0].imshow(img)
ax[0].set_title("Original Image")
ax[1].imshow(img_gray,cmap="gray")
ax[1].set_title("Gray-scaled Image")
ax[2].imshow(img_edge,cmap="gray")
ax[2].set_title("Edge-detected Image")
plt.show()
```

## Corner Detection

```python
## function to calculate the corners in the image
def corner_detection(img, grad_threshold):   #--> requires gray-scaled image as input
    height,width = img.shape
    img_with_corner = np.zeros(img.shape)

    for i in range(height-1):
        for j in range(width-1):
            gradx = abs(int(img[i,j+1]) - int(img[i,j]))  ## grad calculation along the width
            grady = abs(int(img[i+1,j]) - int(img[i,j]))  ## grad calculation along the height
            if gradx >= grad_threshold and grady >= grad_threshold:
                img_with_corner[i,j] = 255
            else:
                img_with_corner[i,j] = 0

    return(img_with_corner.astype(np.uint8))
```

```python
img_path = "../week2/subject01.sad.pgm"
img = plt.imread(img_path)
grad_threshold = 30
img_with_corner = corner_detection(img, grad_threshold)

fig, ax = plt.subplots(1,2,figsize=(12,10))
ax[0].imshow(img,cmap="gray")
ax[0].set_title("Original Image")
ax[1].imshow(img_with_corner,cmap="gray")
ax[1].set_title("Corner-detected Image")
plt.show()
```
